# double bar graph

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python","C","C++","java"]
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("languages",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.title("bar Graph",fontsize=15)


plt.bar(p,y,width,color='r',label="grphical representaion")#linstyle is(:) is doted show
#alpha parameter is bar color in light (transpransy)
plt.bar(p1,z,width,color='y',label="grphical representaion1")

plt.xticks(p+width/2,x,rotation=20)# only three parameter pass,then rotate
plt.legend()
plt.show()
```

# FILL_BETWEEN PLOT IN MATPLOTLIB

import matplotlib.pyplot as plt:

    x=[]
    y=[]
    plt.plot(x,y)
    plt.fill_between()
    plt.show()

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.fill_between(x,y)
plt.plot(x,y,color='r')
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y,color='r')

plt.fill_between(x,y,color="m",where=(x>2)&(x<=4),alpha=0.5)

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.show()
```

# xlim(),ylim()


```python
import matplotlib.pyplot as plt

x=[1,2,3,5]
y=[3,1,5,4]

plt.plot(x,y)

plt.xlim(0,10)
plt.ylim(0,10)



plt.show()
```

# TEXT IN MATPLOTLIB

1. text -------Add text at an arbitrary location of the axes.
2. annotate--- -Add an annotation with an optional arrow at an      arbitrary location of the axes.
3.xlabel---- add a label to the axes's x-axis
4.ylabel---- add a label to the axes's y-axis
5.title------ add a title to the axes


```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)
plt.title("normal graphical representation",fontsize=20)
plt.xlabel("days",fontsize=14)
plt.ylabel("growth",fontsize=20)

plt.text(2,5,"python",fontsize=15,color='m',style="italic",bbox={"facecolor":"red"})


plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)
plt.title("normal graphical representation",fontsize=20)
plt.xlabel("days",fontsize=14)
plt.ylabel("growth",fontsize=20)

plt.text(2,5,"java",fontsize=15,color='m',style="italic",bbox={"facecolor":"red"})


plt.annotate("python",xy=(2,1),xytext=(3,3),arrowprops=dict(facecolor="black",shrink=100))
plt.show()
```
